# 삼각행렬 인덱스 (triu_indixes/ tril_indices)
- 상관계수 행렬, 공분산 행렬, 코사인 유사도 행렬 들은 행과 열이 대칭인 대칭행렬
- 중복을 줄이기 위해서 위쪽 삼각형 또는 아래쪽 삼각형말 다룰 때 사용
* np.triu_indices(n): n x n 행렬에서 위 삼각형의 위치 좌표를 출력
* np.tril_indices(n): n x n 행렬에서 아래 삼각형외 위치 좌표를 출력

```python
mask_up = np.zeros((3,3))
mask_up
```

```python
# 출력은 행, 열 순으로 나옴
row_up, col_up = np.triu_indices(3)
print(row_up)
print(col_up)
```

```python
mask_up[row_up, col_up] = 1
```

```python
mask_up
```

```python
row_bottom, col_bottom = np.tril_indices(3)
```

```python
mask_up[row_bottom, col_bottom] = 3
```

```python
mask_up
```
